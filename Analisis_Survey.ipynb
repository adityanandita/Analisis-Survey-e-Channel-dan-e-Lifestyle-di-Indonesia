# Uji Statistik dan Hipotesis

```python
#Menghitung confidence interval 95% dan Uji Hipotesis
import math
def CI95(kolom, var):
    sampel30 = df[kolom].sample(n = 30, random_state = 1)
    sampel100 = df[kolom].sample(n = 100, random_state = 1)
    Z=1.95996
    print("Confidence Interval 95% untuk variabel " + var + " pada " + kolom + " adalah:")
    
    hitung30 = 0
    for i in range (0, len(sampel30)):
        if sampel30.iloc[i] == var:
            hitung30 += 1
    z30 = Z*math.sqrt(((hitung30/30))*((30-hitung30)/30)/30)
    LCI30 = (hitung30/30) - z30
    UCI30 = (hitung30/30) + z30
    print("n = 30:") 
    print("{:.7f}".format(LCI30), "< Proporsi", var, "<", "{:.7f}".format(UCI30))
    
    hitung100 = 0
    for i in range (0, len(sampel100)):
        if sampel100.iloc[i] == var:
            hitung100 += 1
    z100 = Z*math.sqrt(((hitung100/100)*((100-hitung100)/100))/100)
    LCI100 = (hitung100/100) - z100
    UCI100 = (hitung100/100) + z100
    print("n = 100:") 
    print("{:.7f}".format(LCI100), "< Proporsi", var, "<", "{:.7f}".format(UCI100))
    
    total = 0
    for i in range (0, len(df[kolom])):
        if df[kolom].iloc[i] == var:
            total += 1
    z_all = Z*math.sqrt(((total/len(df[kolom]))*((len(df[kolom])-total)/len(df[kolom])))/len(df[kolom]))
    LCI_all = (total/len(df[kolom])) - z_all
    UCI_all = (total/len(df[kolom])) + z_all
    print("n = ", len(df[kolom]), ":") 
    print("{:.7f}".format(LCI_all), "< Proporsi", var, "<", "{:.7f}".format(UCI_all))
```

```python
#Uji hipotesis untuk rata-rata 1 arah kecil dari
def UH1(kolom, var, p):
    sampel30 = df[kolom].sample(n = 30, random_state = 1)
    sampel100 = df[kolom].sample(n = 100, random_state = 1)
    Z=1.644 #Z untuk alfa 0.05
    q = 1-p
    print("Untuk H1 <", p, ", maka:")
    
    hitung30 = 0
    for i in range (0, len(sampel30)):
        if sampel30.iloc[i] == var:
            hitung30 += 1
    Z_score30 = (((hitung30/30)-p)*30)/(math.sqrt(30*p*q))
    print("Z score untuk 30 data random:")
    print(Z_score30)
    if Z_score30 < (Z*(-1)):
        print("Untuk 30 data random, H0 ditolak")
    else:
        print("Untuk 30 data random, H0 gagal ditolak")
    
    hitung100 = 0
    for i in range (0, len(sampel100)):
        if sampel100.iloc[i] == var:
            hitung100 += 1
    Z_score100 = (((hitung100/100)-p))*100/(math.sqrt(100*p*q))
    print("Z score untuk 100 data random:")
    print(Z_score100)
    if Z_score100 < (Z*(-1)):
        print("Untuk 100 data random, H0 ditolak")
    else:
        print("Untuk 100 data random, H0 gagal ditolak")
    
    total = 0
    for i in range (0, len(df[kolom])):
        if df[kolom].iloc[i] == var:
            total += 1
    Z_score_all = (((total/len(df[kolom]))-p)*len(df[kolom]))/(math.sqrt(len(df[kolom])*p*q))
    print("Z score untuk semua data:")
    print(Z_score_all)
    if Z_score_all < (Z*(-1)):
        print("Untuk semua data, H0 ditolak")
    else:
        print("Untuk semua data, H0 gagal ditolak")
```

```python
#Uji hipotesis untuk rata-rata dengan analisis 2 arah
def UH3(kolom, var, p):
    sampel30 = df[kolom].sample(n = 30, random_state = 1)
    sampel100 = df[kolom].sample(n = 100, random_state = 1)
    Z=1.95996
    q = 1-p
    print("Untuk H1 !=", p, ", maka:")
    
    hitung30 = 0
    for i in range (0, len(sampel30)):
        if sampel30.iloc[i] == var:
            hitung30 += 1
    Z_score30 = (((hitung30/30)-p)*30)/(math.sqrt(30*p*q))
    print("Z score untuk 30 data random:")
    print(Z_score30)
    if Z_score30 > (Z) or Z_score30 < (Z*(-1)):
        print("Untuk 30 data random, H0 ditolak")
    else:
        print("Untuk 30 data random, H0 gagal ditolak")
    
    hitung100 = 0
    for i in range (0, len(sampel100)):
        if sampel100.iloc[i] == var:
            hitung100 += 1
    Z_score100 = (((hitung100/100-p))*100)/(math.sqrt(100*p*q))
    print("Z score untuk 100 data random:")
    print(Z_score100)
    if Z_score100 > (Z) or Z_score100 < (Z*(-1)):
        print("Untuk 100 data random, H0 ditolak")
    else:
        print("Untuk 100 data random, H0 gagal ditolak")
    
    total = 0
    for i in range (0, len(df[kolom])):
        if df[kolom].iloc[i] == var:
            total += 1
    Z_score_all = (((total/len(df[kolom]))-p)*len(df[kolom]))/(math.sqrt(len(df[kolom])*p*q))
    print("Z score untuk semua data:")
    print(Z_score_all)
    if Z_score_all > (Z) or Z_score_all < (Z*(-1)):
        print("Untuk semua data, H0 ditolak")
    else:
        print("Untuk semua data, H0 gagal ditolak")
```
